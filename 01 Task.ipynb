```python
#!/usr/bin/env python
# coding: utf-8

# In[49]:


import random
import statistics as st
import math

n = int(input('  '))
a=[random.gauss(0,math.sqrt(2)) for i in range(n)] #,   

print('Median ', st.median(a))

print('Mean ',st.mean(a))

print('Variance',st.variance(a))

c=[math.sin(i) for i in a]
print('sin X', sum(c[:])/len(c))

arr=[]
for i in a:
    if i>0:
        arr.append(i)
print(arr)

s=0
for i in range(n):
    if s>3:
        print(i)
        break
    s+=a[i]
if(s<=3):
    print(n+1)


# In[ ]:
```
