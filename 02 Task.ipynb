```python
priem = pd.read_csv('Priem.csv')

print('\n#1')
priem1 = priem.loc[:,['Пол', 'Сумма.баллов']]
#priem1.set_index(['Пол','Сумма.баллов'])

priem2 = priem1.groupby(['Пол']).mean()
print(priem2)

print('\n#2')
priem11 = priem.assign(Attestat = priem['Аттестат.с.отличием.оценка'] == 5)
priem22 = priem11.loc[:, ['Attestat', 'Сумма.баллов']]
priem33 = priem22.groupby(['Attestat']).mean()
print(priem33)

print('\n#3')

priem4 = priem11.loc[:, ['Пол', 'Attestat', 'Сумма.баллов']]
priem4 = priem4.groupby(['Пол', 'Attestat']).mean()
print(priem4)

# А КАК СДЕЛАТЬ ТАК ЧТОБЫ БЫЛО 4 СТРОЧКИ???

print('\n#4')

priem5 = priem.loc[priem['ЕГЭ...матем.'] < 90]
priem5 = priem5.loc[priem5['ЕГЭ...матем.'] > 80]
print(priem5['ЕГЭ...матем.'].count())

print('\n#5')
x = 70
y = 80

priem6 = priem.loc[priem['ЕГЭ...матем.'] < x]
priem6 = priem6['ЕГЭ...матем.'].count()

priem7 = priem.loc[priem['ЕГЭ...матем.'] >= x]
priem7 = priem7.loc[priem7['ЕГЭ...матем.'] < y]
priem7 = priem7['ЕГЭ...матем.'].count()

priem8 = priem.loc[priem['ЕГЭ...матем.'] >= y]
priem8 = priem8['ЕГЭ...матем.'].count()

priempriem = pd.DataFrame({'<x': [priem6],
                          'x<=...<y': [priem7],
                          '>=y': [priem8]})
print(priempriem)
```
