```python
#№1

import scipy.stats as sci
import math as ma
import statistics as stat
import numpy as np
import matplotlib.pyplot as plt

n = 7
fig, ax = plt.subplots(figsize=(10, 6))

#plt.figure(1)
#x = np.linspace(-50, 50, 1)
for k in range(0, n):
    x = np.linspace(-20, 20, 1000)
    y = sci.norm.pdf(x, loc = 0, scale = k+1)
    ax.plot(x, y, label=f'k={k}')
    
#ax.legend()
ax.set_xlabel('Значение')
ax.set_ylabel('Плотность')
    
plt.title('Графики плотностей N(1, n) для разных значений k')    
plt.grid(True)

plt.show()
    
```

```python
#№9*

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import ipywidgets as widgets
from IPython.display import display

# Создаем функцию для построения графика
def plot_density(n):
    fig, ax = plt.subplots(figsize=(10, 6))

    #plt.figure(1)
    #x = np.linspace(-50, 50, 1)
    for k in range(0, n):
        x = np.linspace(-25, 25, 1000)
        y = sci.norm.pdf(x, loc = 0, scale = k+1)
        ax.plot(x, y, label=f'k={k}')
    
    #ax.legend()
    ax.set_xlabel('Значение')
    ax.set_ylabel('Плотность')
    
    plt.title('Графики плотностей N(1, n) для разных значений k')    
    plt.grid(True)


    # Покажем график
    plt.show()

# Создаем ползунок для изменения значения n
n_slider = widgets.IntSlider(min=1, max=20, step=1, value=7)

# Создаем интерактивное отображение графика
interactive_plot = widgets.interactive(plot_density, n=n_slider)

# Отображаем ползунок и график
display(interactive_plot)
```

```python
#№2

import numpy as np
import matplotlib.pyplot as plt

# Задаём параметры для случайных величин
np.random.seed(0)  # Для воспроизводимости результатов
n_simulations = 1000

# Моделирование случайных величин X ~ Bin(100, 0.04) и Y ~ Poisson(4)
X1 = np.random.binomial(100, 0.04, n_simulations)
Y1 = np.random.poisson(4, n_simulations)

# Моделирование случайных величин X ~ Bin(1000, 0.004) и X ~ Bin(10, 0.4)
X2 = np.random.binomial(1000, 0.004, n_simulations)
X3 = np.random.binomial(10, 0.4, n_simulations)

# Создаём фигуры и оси для графика
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

# Постройте столбцовые диаграммы для X ~ Bin(100, 0.04) и Y ~ Poisson(4)
axs[0, 0].hist(X1, bins=20, color='blue', alpha=0.6, label='X ~ Bin(100, 0.04)')
axs[0, 0].set_title('X ~ Bin(100, 0.04)')
axs[0, 0].legend()
axs[0, 1].hist(Y1, bins=20, color='red', alpha=0.6, label='Y ~ Poisson(4)')
axs[0, 1].set_title('Y ~ Poisson(4)')
axs[0, 1].legend()

# Постройте столбцовые диаграммы для X ~ Bin(1000, 0.004) и X ~ Bin(10, 0.4)
axs[1, 0].hist(X2, bins=20, color='green', alpha=0.6, label='X ~ Bin(1000, 0.004)')
axs[1, 0].set_title('X ~ Bin(1000, 0.004)')
axs[1, 0].legend()
axs[1, 1].hist(X3, bins=20, color='purple', alpha=0.6, label='X ~ Bin(10, 0.4)')
axs[1, 1].set_title('X ~ Bin(10, 0.4)')
axs[1, 1].legend()

# Постройте сравнительные графики
axs[0, 2].hist(X1, bins=20, color='blue', alpha=0.6, label='X ~ Bin(100, 0.04)')
axs[0, 2].hist(Y1, bins=20, color='red', alpha=0.6, label='Y ~ Poisson(4)')
axs[0, 2].set_title('Сравнение X и Y')
axs[0, 2].legend()

axs[1, 2].hist(X2, bins=20, color='green', alpha=0.6, label='X ~ Bin(1000, 0.004)')
axs[1, 2].hist(X3, bins=20, color='purple', alpha=0.6, label='X ~ Bin(10, 0.4)')
axs[1, 2].set_title('Сравнение X и X')
axs[1, 2].legend()

# Настройте отображение графиков
plt.tight_layout()

# Покажите графики
plt.show()
```

```python
#№3
import numpy as np
#import mathplot as math
import pandas as pd
priem = pd.read_csv('Priem.csv')

# Создаём фигуры и оси для графика
fig, axs = plt.subplots(2, 2, figsize=(15, 10))

X1 = priem['ЕГЭ...рус.яз.']
X2 = priem['ЕГЭ...матем.']

# Постройте столбцовые диаграммы для X1 - математика; X2 - русский
axs[0, 0].hist(X1, bins=20, color='green', alpha=0.6, label='Математика')
axs[0, 0].set_title('Математика')
axs[0, 0].legend()
axs[0, 1].hist(X2, bins=20, color='blue', alpha=0.6, label='Русский')
axs[0, 1].set_title('Русский')
axs[0, 1].legend()

# Постройте сравнительные графики
axs[1, 0].hist(X1, bins=20, color='green', alpha=0.6, label='Математика')
axs[1, 0].hist(X2, bins=20, color='blue', alpha=0.6, label='Русский')
axs[1, 0].set_title('Сравнение X1 и X2')
axs[1, 0].legend()

# Настройте отображение графиков
plt.tight_layout()

# Покажите графики
plt.show()

```
