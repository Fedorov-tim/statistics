```python
#1

import scipy.stats as sci
import math as ma
import statistics as stat
import numpy as np
import matplotlib.pyplot as plt

n = 7
fig, ax = plt.subplots(figsize=(10, 6))

#plt.figure(1)
#x = np.linspace(-50, 50, 1)
for k in range(0, n):
    x = np.linspace(-20, 20, 1000)
    y = sci.norm.pdf(x, loc = 0, scale = k+1)
    ax.plot(x, y, label=f'k={k}')
    
#ax.legend()
ax.set_xlabel('')
ax.set_ylabel('')
    
plt.title('  N(1, n)    k')    
plt.grid(True)

plt.show()
    
```

```python
#9*

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import ipywidgets as widgets
from IPython.display import display

#     
def plot_density(n):
    fig, ax = plt.subplots(figsize=(10, 6))

    #plt.figure(1)
    #x = np.linspace(-50, 50, 1)
    for k in range(0, n):
        x = np.linspace(-25, 25, 1000)
        y = sci.norm.pdf(x, loc = 0, scale = k+1)
        ax.plot(x, y, label=f'k={k}')
    
    #ax.legend()
    ax.set_xlabel('')
    ax.set_ylabel('')
    
    plt.title('  N(1, n)    k')    
    plt.grid(True)


    #  
    plt.show()

#      n
n_slider = widgets.IntSlider(min=1, max=20, step=1, value=7)

#    
interactive_plot = widgets.interactive(plot_density, n=n_slider)

#    
display(interactive_plot)
```

```python
#2

import numpy as np
import matplotlib.pyplot as plt

#     
np.random.seed(0)  #   
n_simulations = 1000

#    X ~ Bin(100, 0.04)  Y ~ Poisson(4)
X1 = np.random.binomial(100, 0.04, n_simulations)
Y1 = np.random.poisson(4, n_simulations)

#    X ~ Bin(1000, 0.004)  X ~ Bin(10, 0.4)
X2 = np.random.binomial(1000, 0.004, n_simulations)
X3 = np.random.binomial(10, 0.4, n_simulations)

#      
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

#     X ~ Bin(100, 0.04)  Y ~ Poisson(4)
axs[0, 0].hist(X1, bins=20, color='blue', alpha=0.6, label='X ~ Bin(100, 0.04)')
axs[0, 0].set_title('X ~ Bin(100, 0.04)')
axs[0, 0].legend()
axs[0, 1].hist(Y1, bins=20, color='red', alpha=0.6, label='Y ~ Poisson(4)')
axs[0, 1].set_title('Y ~ Poisson(4)')
axs[0, 1].legend()

#     X ~ Bin(1000, 0.004)  X ~ Bin(10, 0.4)
axs[1, 0].hist(X2, bins=20, color='green', alpha=0.6, label='X ~ Bin(1000, 0.004)')
axs[1, 0].set_title('X ~ Bin(1000, 0.004)')
axs[1, 0].legend()
axs[1, 1].hist(X3, bins=20, color='purple', alpha=0.6, label='X ~ Bin(10, 0.4)')
axs[1, 1].set_title('X ~ Bin(10, 0.4)')
axs[1, 1].legend()

#   
axs[0, 2].hist(X1, bins=20, color='blue', alpha=0.6, label='X ~ Bin(100, 0.04)')
axs[0, 2].hist(Y1, bins=20, color='red', alpha=0.6, label='Y ~ Poisson(4)')
axs[0, 2].set_title(' X  Y')
axs[0, 2].legend()

axs[1, 2].hist(X2, bins=20, color='green', alpha=0.6, label='X ~ Bin(1000, 0.004)')
axs[1, 2].hist(X3, bins=20, color='purple', alpha=0.6, label='X ~ Bin(10, 0.4)')
axs[1, 2].set_title(' X  X')
axs[1, 2].legend()

#   
plt.tight_layout()

#  
plt.show()
```
